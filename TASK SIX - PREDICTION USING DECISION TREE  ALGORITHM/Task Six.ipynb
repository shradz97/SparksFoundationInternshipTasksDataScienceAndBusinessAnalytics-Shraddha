<h1>Import the libraries</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix 
from sklearn.tree import plot_tree
```

<h1>Import the dataset</h1>

```python
df=pd.read_csv("Iris.csv")
```

<h1>Displaying first five rows of the dataset</h1>


```python
df.head()
```

<h1>Data Processing</h1>

```python
X = df.iloc[:, 1:5].values
Y = df.iloc[:, 5].values
```

<h3>Splitting the dataset</h3>

```python
X_train, X_test, Y_train, Y_test = train_test_split(X , Y, test_size = 0.2, random_state = 42)
print("Training split input- ", X_train.shape)
print("Testing split input- ", X_test.shape)
```

<h3>Assigning and fitting the data</h3>

```python
dtree=DecisionTreeClassifier()
dtree.fit(X_train,Y_train)
```

<h3>Predict</h3>

```python
Y_pred = dtree.predict(X_test)
Y_pred 
```

<h1>Visualization</h1>

```python
plt.figure(figsize = (20,10))
dec_tree = plot_tree(decision_tree=dtree, feature_names = df.columns, 
                     class_names =["setosa", "vercicolor", "verginica"] , filled = True , precision = 4, rounded = True)
```
