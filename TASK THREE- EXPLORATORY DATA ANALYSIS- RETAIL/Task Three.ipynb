<h1>Data Processing</h1><br>
<h3>Dropping duplicate values from the dataset</h3>

```python
retail.drop_duplicates(subset=None,keep='first',inplace=True)
retail.duplicated().sum()
```

<h1>PLOTS</h1>

<h3>Plot 1: Category vs Discount (Scatter Plot)</h3>

```python
retail.plot.scatter('Category','Discount')
```

<h3>Plot 2: Category vs Profit (Scatter Plot)</h3>

```python
retail.plot.scatter('Category','Profit')
```

<h3>Plot 3: Category vs Sales (Scatter Plot)</h3>

```python
retail.plot.scatter('Category','Sales')
```

<h3>Plot 4: State vs Total Sales (Bar Graph)</h3>

```python
Sales = retail.groupby("State").Sales.sum()
plt.style.use('seaborn')
Sales.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Sales",fontsize =13)
plt.title("States by Sales",fontsize =16)
plt.show()
```

<h3>Plot 5: State vs Profit (Bar Graph)</h3>

```python
Profit = retail.groupby("State").Profit.sum()
plt.style.use('seaborn')
Profit.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Profit",fontsize =13)
plt.title("Profit in each state",fontsize =16)
plt.show()
```

<h3>Plot 6: Ship Mode vs Postal Code (Bar Graph)</h3>

```python
Profit = retail.groupby("ShipMode").PostalCode.sum()
plt.style.use('seaborn')
Profit.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("ShipMode", fontsize =13)
plt.ylabel("PostalCode",fontsize =13)
plt.show()
```

<h3>Plot 7: Discount Categorised by Region (Pie Chart)</h3>

```python
Profit = retail.groupby('Region').Discount.sum()
plt.style.use('seaborn')
Profit.plot(kind ='pie', figsize =(7,8), fontsize =14)
plt.title("Discount categorised by region",fontsize =16)
plt.show()
```

<h3>Plot 8: Sales Categorised by Region (Pie Chart)</h3>

```python
Profit = retail.groupby('Region').Sales.sum()
plt.style.use('seaborn')
Profit.plot(kind ='pie', figsize =(7,8), fontsize =14)
plt.title("Sales categorised by region",fontsize =16)
plt.show()
```

<h3>Plot 9: Profit Categorised by Region (Pie Chart)</h3>

```python
Profit = retail.groupby('Region').Profit.sum()
plt.style.use('seaborn')
Profit.plot(kind ='pie', figsize =(7,8), fontsize =14)
plt.title("Profit categorised by region",fontsize =16)
plt.show()
```

<h3>Plot 10: Region categorised by Quantity (Pie Chart)</h3>

```python
Profit = retail.groupby('Region').Quantity.sum()
plt.style.use('seaborn')
Profit.plot(kind ='pie', figsize =(7,8), fontsize =14)
plt.title("Region categorised by Quantity",fontsize =16)
plt.show()
```

<h3>Plot 11: Category Categorised by Quantity(Pie Chart)</h3>

```python
Profit = retail.groupby('Category').Quantity.sum()
plt.style.use('seaborn')
Profit.plot(kind ='pie', figsize =(7,8), fontsize =14)
plt.title("Category categorised by Quantity",fontsize =16)
plt.show()
```

<h3>Plot 12: Sub-Category categorised by Sales (Pie Chart)</h3>

```python
Profit = retail.groupby('SubCategory').Sales.sum()
plt.style.use('seaborn')
Profit.plot(kind ='pie', figsize =(10,8), fontsize =14)
plt.title("Sub-Category categorised by Sales",fontsize =16)
plt.show()
```

<h3>Plot 13: Sub-Category categorised by Discount(Pie Chart)</h3>

```python
Profit = retail.groupby('SubCategory').Discount.sum()
plt.style.use('seaborn')
Profit.plot(kind ='pie', figsize =(14,8), fontsize =14)
plt.title("Sub-Category categorised by Discount",fontsize =16)
plt.show()
```

<h3>Plot 14: Profit vs Sales for Category represented according to discount (Relplot)</h3>

```python
sns.set_theme(style="white")
sns.relplot(x="Profit", y="Sales", hue="Category", size="Discount",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=8, data=retail)
```

<h3>Plot 15: Correlation Heatmap</h3>

```python
plt.figure(figsize=(7,5))
# Store heatmap object in a variable to easily access it when you want to include more features (such as title).
# Set the range of values to be displayed on the colormap from -1 to 1, and set the annotation to True to display the correlation values on the heatmap.
heatmap = sns.heatmap(retail.corr(), vmin=-1, vmax=1, annot=True)
# Give a title to the heatmap. Pad defines the distance of the title from the top of the heatmap.
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

<h3>Plot 18: Pairplot</h3>

<h3>Plot 19: Sales vs Profit vs Quantity(Hexagonal Binning Plot)</h3>
