<h1>Import the libraries</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
%matplotlib inline
```

<h1>Import the dataset</h1>

```python
df=pd.read_csv('ProjectOneData.csv')
df1=pd.read_csv('testcase.csv')
```

<h1>Displaying first five rows of the dataset</h1>

```python
df.head()
```

<h1>Plotting the dataset</h1>

<h3>Line Plot</h3>

```python
df.plot()
```

<h3>Scatter Plot</h3>

```python
df.plot.scatter('Hours','Scores')
plt.title('Scatter Plot')
```

<h1>Data Processing</h1>

```python
X = df.iloc[:, :-1].values  
Y = df.iloc[:, 1].values
```

<h3>Splitting the dataset</h3>

```python
x_train, x_test,y_train,y_test = train_test_split(X,Y,test_size =0.2)
x_train
```

<h1>Linear Regression</h1>

<h3>Importing the model and assigning a variable</h3>

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression()
```

<h3>Fitting the data</h3>

```python
clf.fit(x_train,y_train)
```

<h3>Predict</h3>

```python
clf.predict(x_test)
```

<h3>Accuracy</h3>

```python
acc=clf.score(x_test,y_test)
acc
```

<h3>Plotting</h3>

```python
line = clf.coef_*X+clf.intercept_

# Plotting for the test data
plt.scatter(X, Y)
plt.plot(X, line);
plt.show()
```

<h3>Predicting using given testcase</h3>

```python
own_pred = clf.predict(df1)
print("No of Hours = {}".format(df1))
print("Predicted Score = {}".format(own_pred[0]))
```
